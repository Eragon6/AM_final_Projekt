```python
import numpy as np
import matplotlib.pyplot as plt

def logistic(r, x):
    return r*x*(1-x)

n = 10000
r = np.linspace(2.5, 4.0, n)
iterations = 1000
last = 100
x = 1e-5 * np.ones(n)
Feigenbaum = np.zeros(n)

for i in range(iterations):
    x = logistic(r, x)
    Feigenbaum += np.log(abs(r - 2*r*x))
    if i >= (iterations - last):
        plt.plot(r, x, ",k", alpha = 0.02)

plt.xlim(2.5, 4.0)
plt.title("Feigenbaum-Diagramm")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iterations):
    z = 0
    for i in range(max_iterations):
        z = z**2 + c
        if abs(z) > 2:
            return i
    return max_iterations

# Parameter des Startfraktals
xmin, xmax = -2.5, 1.5
ymin, ymax = -1.8, 1.8
width, height = 800, 800
max_iterations = 100

# Erstellen des Koordinatengitters

x = np.linspace(xmin, xmax, width)
y = np.linspace(ymin, ymax, height)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Berechnung fraktaler Werte
fractal = np.zeros((height, width))
for i in range(height):
    for j in range(width):
        fractal[i, j] = mandelbrot(Z[i, j], max_iterations)

# ANeige der Fraktal
plt.figure(figsize=(10, 10))
plt.imshow(fractal, cmap='hot', extent=(xmin, xmax, ymin, ymax))
plt.colorbar(label='Iterations')
plt.title('Fractal de Mandelbrot')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.show()
```
